```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, FFMpegWriter
from matplotlib.ticker import FuncFormatter
from IPython.display import HTML

# -----------------------------
# Data
# -----------------------------
categories = [
    "Strategie und Purpose*",
    "Spirit und Agilität",
    "Mensch im Fokus*",
    "Nährboden für Wachstum"
]

data_2023 = np.array([76.3, 72.1, 72.1, 80.4])
data_2025 = np.array([78.8, 73.5, 74.9, 82.7])

# Replace with exact values if you have them
global_vals = np.array([77.5, 69.0, 69.0, 80.0])

# -----------------------------
# Animation settings
# -----------------------------
fps = 30
duration_sec = 8
pause_duration_sec = 2

frames = fps * duration_sec
pause_frames = fps * pause_duration_sec
total_frames = frames + pause_frames

# -----------------------------
# Layout
# -----------------------------
x = np.arange(len(categories))
bar_width = 0.28

fig, ax = plt.subplots(figsize=(10, 5))
fig.patch.set_facecolor("white")
ax.set_facecolor("white")

# -----------------------------
# Colors
# -----------------------------
color_2023 = "#B7B7B7"
color_2025 = "#FF7A00"
color_global = "#FF4D4D"
label_color = "#7A7A7A"
grid_color = "#D9D9D9"

# -----------------------------
# Axes styling
# -----------------------------
ax.set_ylim(0, 105)
ax.set_yticks(np.arange(0, 101, 10))
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f"{int(y)}"))
ax.set_xticks(x)
ax.set_xticklabels(categories, fontsize=8, color=label_color)

ax.grid(axis="y", color=grid_color, linewidth=1, alpha=0.8)
ax.set_axisbelow(True)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_color(grid_color)

ax.tick_params(axis="y", length=0, colors="#9A9A9A", labelsize=11)
ax.tick_params(axis="x", colors=label_color, labelsize=8)

# -----------------------------
# Bars start at zero for animation
# -----------------------------
bars_2023 = ax.bar(
    x - bar_width / 2,
    np.zeros_like(data_2023),
    width=bar_width,
    color=color_2023,
    edgecolor=color_2023,
    label="2023"
)

bars_2025 = ax.bar(
    x + bar_width / 2,
    np.zeros_like(data_2025),
    width=bar_width,
    color=color_2025,
    edgecolor=color_2025,
    label="2025"
)

# Static global line
(line_global,) = ax.plot(
    x,
    global_vals,
    linestyle=(0, (3, 3)),
    linewidth=2,
    color=color_global,
    label="Global"
)

# -----------------------------
# Value labels above bars
# -----------------------------
def de_number(v):
    return f"{v:.1f}".replace(".", ",")

labels_2023 = []
labels_2025 = []

for i in range(len(categories)):
    t1 = ax.text(
        x[i] - bar_width / 2,
        0,
        "",
        ha="center",
        va="bottom",
        fontsize=10,
        color=label_color,
        fontweight="bold"
    )
    t2 = ax.text(
        x[i] + bar_width / 2,
        0,
        "",
        ha="center",
        va="bottom",
        fontsize=10,
        color=label_color,
        fontweight="bold"
    )
    labels_2023.append(t1)
    labels_2025.append(t2)

# -----------------------------
# Legend
# -----------------------------
ax.legend(
    loc="upper center",
    bbox_to_anchor=(0.5, -0.12),
    ncol=3,
    frameon=False,
    fontsize=12,
    handlelength=2.0,
    columnspacing=1.5,
    handletextpad=0.4
)

# -----------------------------
# Animation
# -----------------------------
def animate(i):
    if i >= frames:
        progress = 1.0
    else:
        progress = i / (frames - 1)

    current_2023 = data_2023 * progress
    current_2025 = data_2025 * progress

    # Update bars
    for j, bar in enumerate(bars_2023):
        bar.set_height(current_2023[j])

    for j, bar in enumerate(bars_2025):
        bar.set_height(current_2025[j])

    # Update labels
    for j in range(len(categories)):
        labels_2023[j].set_y(current_2023[j] + 1.2)
        labels_2023[j].set_text(de_number(current_2023[j]) if progress > 0 else "")
        labels_2023[j].set_alpha(min(1, progress * 3))

        labels_2025[j].set_y(current_2025[j] + 1.2)
        labels_2025[j].set_text(de_number(current_2025[j]) if progress > 0 else "")
        labels_2025[j].set_alpha(min(1, progress * 3))

    return [*bars_2023, *bars_2025, *labels_2023, *labels_2025, line_global]

anim = FuncAnimation(
    fig,
    animate,
    frames=total_frames,
    interval=1000 / fps,
    blit=False
)

# -----------------------------
# Save MP4
# -----------------------------
mp4_path = "performance_zeitvergleich_static_line.mp4"
writer = FFMpegWriter(
    fps=fps,
    codec="libx264",
    bitrate=2500,
    extra_args=["-pix_fmt", "yuv420p"]
)
anim.save(mp4_path, writer=writer, dpi=200)

print(f"Saved MP4: {mp4_path}")

plt.close(fig)
HTML(anim.to_html5_video())
```
